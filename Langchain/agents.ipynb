```python
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
load_dotenv()
model=init_chat_model("groq:llama-3.1-8b-instant")

response= model.invoke("Write an essay on AI")
```

```python
response
```

```python
# tools 

from langchain.tools import tool

@tool
def get_weather(location:str)->str:
    """Get the weather at a location """
    return f"Its sunny in {location}"

model_with_tools= model.bind_tools([get_weather])
```

```python
response = model_with_tools.invoke("What is weather in Banglore")

print(response)
```

```python
response.tool_calls
```

# Tool execution Loop

```python
# Step 1 : Model generates Tool calls


messages = [{"role": "user" , "content": "What is weather in Banglore"}]
ai_mesage = model_with_tools.invoke(messages)
messages.append(ai_mesage)

#Step 2 execute tools and collect results 

for tool_call in ai_mesage.tool_calls:
    #execute the tool with generated arguments
    tool_result = get_weather.invoke(tool_call)
    messages.append(tool_result)

# Step 3 : Pass result back to model 

final_response = model_with_tools.invoke(messages)
print(final_response.text)
```
