```python
from langgraph.graph import StateGraph  , START , END
from dotenv import load_dotenv
from typing import TypedDict , Annotated
from langchain_google_genai import ChatGoogleGenerativeAI
from pydantic import BaseModel , Field
import operator
```

```python
load_dotenv()
```

```python
model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=1.0,
    max_tokens=None,
    timeout=None,
    max_retries=2,
)
```

```python
class EvaluationSchema(BaseModel):

    feedback :str = Field(description = 'detailed feedback for the essay ')
    score : int = Field(description = 'Score out of 10' , ge=0 , le=10 )
```

```python
structured_model = model.with_structured_output(EvaluationSchema)
```

```python
essay = """ 
The Rise of DHH Culture in India and the Global Outshining of Indian Rap
Introduction

Over the last decade, India has witnessed a cultural shift that few anticipated but many now celebrate—the meteoric rise of Desi Hip Hop (DHH). What was once an underground movement limited to niche listeners has transformed into a powerful cultural force shaping youth identity, social discourse, and global perceptions of Indian music. Today, Indian rap is no longer trying to “catch up” with the West; instead, it is standing alongside global hip-hop movements with its own voice, language, and truth.

Origins of Desi Hip Hop: From Streets to Stories

DHH did not emerge from luxury studios or commercial boardrooms. It was born on the streets, chawls, college campuses, and small recording rooms where young artists used rap as a medium to express frustration, ambition, inequality, and identity.

Early pioneers like Bohemia laid the foundation by introducing Punjabi rap to mainstream audiences. Later, artists such as Naezy, DIVINE, Emiway Bantai, Raftaar, and KR$NA expanded the movement by rapping in Hindi, regional languages, and urban slang—making hip hop feel local, raw, and relatable.

Unlike Western rap, which often centers on excess and individualism, Indian hip hop leaned heavily on:

Class struggle

Social mobility

Systemic inequality

Personal survival stories

This authenticity became DHH’s strongest currency.

Language as Power: Redefining Global Hip Hop Norms

One of the most revolutionary aspects of DHH is its rejection of linguistic conformity. Indian rappers did not abandon their languages to fit global standards. Instead, they exported Hindi, Punjabi, Urdu, Marathi, Tamil, and Hinglish rap to the world.

This linguistic confidence sent a clear message:

You don’t need English validation to have global relevance.

Much like Latin trap and Korean hip hop, DHH proved that emotion, rhythm, and truth transcend language barriers. International listeners may not understand every word, but they feel the intensity, hunger, and realism embedded in Indian rap.

Digital Platforms: The Democratization of Stardom

The rise of DHH is inseparable from the rise of YouTube, Spotify, Instagram, and independent streaming platforms. Traditional gatekeepers—music labels, radio stations, and film industries—lost their monopoly.

Artists could now:

Release music independently

Build cult fanbases organically

Monetize without compromise

Speak without censorship

This digital freedom allowed Indian rappers to challenge not just musical norms, but power structures within the industry itself. Diss tracks, independent drops, and viral freestyles became tools of influence and disruption.

Cultural Legitimacy: From “Noise” to Narrative

For years, rap in India was dismissed as aggressive, westernized, or “not real music.” That perception has radically changed. DHH is now recognized as:

A voice of the urban youth

A mirror of social realities

A legitimate art form

Films like Gully Boy may have popularized the movement, but the culture had already built its foundation. What cinema did was amplify a truth that already existed.

Today, rap lyrics are quoted in classrooms, debated on social media, and studied as cultural texts.

India on the Global Rap Map

Indian hip hop is no longer invisible on the world stage. Collaborations with international artists, global reaction videos, festival appearances, and worldwide streaming numbers have placed India firmly in the global rap conversation.

What makes India stand out globally is:

Diversity of sound (regional flows, folk fusion, classical elements)

Narrative depth (stories of survival, resistance, ambition)

Cultural layering (history, caste, class, migration, identity)

India doesn’t imitate global rap—it adds a new chapter to it.

Beyond Music: A Cultural Movement

DHH is no longer just music. It is:

A fashion statement

A political voice

A social movement

A form of self-assertion

It empowers young Indians to say:

My story matters, even if the system ignores me.

That is why DHH resonates so deeply—it gives dignity to voices that were historically unheard.

Conclusion

The rise of Desi Hip Hop is not a trend—it is a cultural awakening. Indian rap has moved from the margins to the mainstream, from imitation to innovation, and from local struggles to global relevance.

As Indian artists continue to tell their stories unapologetically, the world is not just listening—it is learning.

India is no longer knocking on the global hip-hop door.
It has already entered—and it’s changing the sound inside.
"""
```

```python
prompt = f' Evauluate the language quality of the following essay and provide a feedback and assign a score out of 10 \n {essay}'
structured_model.invoke(prompt)
```

```python
structured_model.invoke(prompt).feedback
```

```python
structured_model.invoke(prompt).score
```
