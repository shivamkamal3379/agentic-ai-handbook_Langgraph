What is Streaming ???

In LLMS streaming means the model starts sending tokens words as soon as they are generated instead of waiting for the entire response to be ready before returning it

Why streaming 

Faster response time low drop off rates


Mimics human like conversations(build trust, feels alive, keeps the user engaged )

Important for multi model UI's

Better UX for long output such as code

You can cancel midway saving tokens

You can interleave UI updates -  Example show thinkings, show research tools
